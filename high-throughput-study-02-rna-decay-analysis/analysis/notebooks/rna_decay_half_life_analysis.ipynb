```python
import pandas as pd

# Load the dataset
file_path = "Timecourse1_Interpolated_Data.csv"  # Ensure the path is correct
data = pd.read_csv(file_path)

# Exclude the 'YORF' row 
data_filtered = data[data['Time course #'] != 'YORF']  

# Identify and remove rows where all values are 1
data_filtered = data_filtered.loc[~(data_filtered.iloc[:, 1:] == 1).all(axis=1)]
print(data_filtered.head(7))

total_rows = data_filtered.shape[0]  
print(f"Total rows: {total_rows}")


# Save the cleaned data to a new CSV file
output_file_path = "Cleaned_Timecourse1_Data.csv"
data_filtered.to_csv(output_file_path, index=False)

print("Cleaned_Timecourse1_Data Saved")
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

# Defining the exponential decay function:
def exp_decay(t, k):
    """Exponential decay function."""
    return np.exp(-k * t)

# Update the file path 
file_path = "Cleaned_Timecourse1_Data.csv"

# Loading the data
data = pd.read_csv(file_path)

if 'YORF' in data.iloc[:, 0].values:
    data = data[data.iloc[:, 0] != 'YORF']

# Assuming the first column now correctly represents gene names after 'YORF' exclusion
genes = data.iloc[:, 0].values
time_points = np.array([0, 5, 10, 15, 20, 30, 40, 50, 60])

half_lives_timecourse1 = []

# Iterate over each gene to calculate the half-life
for gene_index in range(len(genes)):
    gene_data = data.iloc[gene_index, 1:].astype(float).values  
    gene_data_timecourse1 = gene_data[:9]  

    try:
        # Fit the exponential decay model to the data
        popt, _ = curve_fit(exp_decay, time_points, gene_data_timecourse1, maxfev=10000)
        half_life = np.log(2) / popt[0]  # Calculate the half-life
        half_lives_timecourse1.append(half_life)
    except RuntimeError as e:
        print(f"Curve fitting failed for gene at index {gene_index} ({genes[gene_index]}): {e}")
        half_lives_timecourse1.append(np.nan)  

# Create a DataFrame to store the results
results_df = pd.DataFrame({
    'Gene': genes,
    'Half-life Time Course 1': half_lives_timecourse1
})

print(results_df.head(11))


# Save the results
output_path = "Half_Lives_TimeCourse1_Results.csv"
results_df.to_csv(output_path, index=False)

total_rows = results_df.shape[0]  
print(f"Total rows: {total_rows}")


print("Half_Lives_TimeCourse1_Results is Successfully Saved")
```

```python
#-------------------------Calculating the Average Half-Life for all the three time course--------------------#
```

```python
import pandas as pd
import numpy as np

def load_and_prepare(file_path):
    """Load CSV data, ensuring gene names are used as the index."""
    data = pd.read_csv(file_path)
    return data

# Paths to the half-life CSV files for each time course
file_path_tc1 = "Half_Lives_TimeCourse1_Results.csv"
file_path_tc2 = "Half_Lives_Timecourse 2_Results.csv"
file_path_tc3 = "Half_Lives_Timecourse3_Results.csv"

# Load the half-life results for each time course
data_tc1 = load_and_prepare(file_path_tc1)
data_tc2 = load_and_prepare(file_path_tc2)
data_tc3 = load_and_prepare(file_path_tc3)

# Merge the half-life results on the 'Gene' column, using outer join to include all genes
merged_data = pd.merge(data_tc1, data_tc2, on='Gene', how='outer', suffixes=('_TC1', '_TC2'))
merged_data = pd.merge(merged_data, data_tc3, on='Gene', how='outer')

# Rename the columns
merged_data.rename(columns={
    'Half-life Time Course 1': 'Half-life_TC1',
    'Half-life Time Course 2': 'Half-life_TC2',
    'Half-life Time Course 3': 'Half-life_TC3'}, inplace=True)

# Calculate the average half-life for each gene
merged_data['Average_Half_Life'] = merged_data[['Half-life_TC1', 'Half-life_TC2', 'Half-life_TC3']].mean(axis=1)

# Save the merged data with average half-lives to a new file
output_path = 'Average_Half_Lives_Results.csv'
merged_data.to_csv(output_path, index=False)
print(merged_data.head(7))

total_rows = merged_data.shape[0]
print(f"Total rows: {total_rows}")

print("Average half-lives results successfully saved")
```
