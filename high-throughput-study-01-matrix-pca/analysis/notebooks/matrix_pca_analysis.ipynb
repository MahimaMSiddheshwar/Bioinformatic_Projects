```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Computing Pearson Correlation between Matrix 1 and Matrix 2

# Read Matrix 1 and Matrix 2 from CSV files
matrix1 = pd.read_csv("PC_Matrix1_data.csv", index_col=0)
matrix2 = pd.read_csv("PC_Matrix2_data.csv", index_col=0)

# Dropped 'Unamed' column which is an irrelevant columns with no values
matrix1 = matrix1.drop(['Unnamed: 13'], axis=1)
matrix2 = matrix2.drop(['Unnamed: 13'], axis=1)

# Computing Pearson correlation between Matrix 1 and Matrix 2
combined_matrix1_matrix2 = pd.DataFrame(index=matrix1.columns, columns=matrix2.columns)
for column1 in matrix1.columns:
    for column2 in matrix2.columns:
        correlation = np.corrcoef(matrix1[column1], matrix2[column2])[0, 1]      # Handle missing or non-numeric values
        
        if np.isnan(correlation):
            correlation = 0
        combined_matrix1_matrix2.loc[column1, column2] = correlation

# Converting data to numeric format
combined_matrix1_matrix2 = combined_matrix1_matrix2.astype(float)

# Creating heatmap for the combined correlation matrix 1 and matrix 2
plt.figure(figsize=(12, 12))
sns.heatmap(combined_matrix1_matrix2, annot=True, cmap='coolwarm', fmt=".3f")
plt.title('Pearson Correlation Between Matrix 1 and Matrix 2')
plt.show()
```
